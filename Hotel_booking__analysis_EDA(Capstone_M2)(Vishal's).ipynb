10. What is the optimal length of stay in order to get the best daily rate

````python
# Chart - 10 visualization code
df_hotel["total_stay"]=df_hotel["stays_in_weekend_nights"] + df_hotel["stays_in_week_nights"]

# Group by hotel type and length of stay, then calculate the average ADR
avg_adr_per_stay=df_hotel.groupby(["hotel", "total_stay"])["adr"].mean().reset_index()

# Plot the results
plt.figure(figsize=(14, 6))
sns.lineplot(data=avg_adr_per_stay, x="total_stay", y="adr", hue="hotel", marker="o")
plt.title("Length of Stay for Different Hotel Types to get best Average Daily Rate")
plt.xlabel("Length of Stay (nights)")
plt.ylabel("Average Daily Rate (ADR)")
plt.legend(title="Hotel Type")
plt.grid(True)
plt.show()
````
